```python
with open('Disgardium_51.txt') as file:
    text = file.readlines()
```

```python
text =  str(text)
```

```python

import re
from random import choice
from collections import defaultdict

r_alphabet = re.compile(u'[а-яА-Я0-9-]+')
```

```python
tokens = [token for token in r_alphabet.findall(text)]
print(tokens)
```

```python
def gen_trigrams(tokens):
    t0, t1 = next(tokens), next(tokens)
    for t2 in tokens:
        yield t0, t1, t2
        t0, t1 = t1, t2
```

```python
tokens = iter(tokens)
my_tri = gen_trigrams(tokens)
print(*my_tri)
```

```python
tokens = iter(tokens)
trigrams = []
t0, t1 = next(tokens), next(tokens)
for t2 in tokens:
    trigrams.append((t0, t1, t2))
    t0, t1 = t1, t2
```

```python
print(trigrams[:15])
```

```python
bi, tri = defaultdict(lambda: 0.0), defaultdict(lambda: 0.0)

for t0, t1, t2 in trigrams:
    bi[t0, t1] += 1
    tri[t0, t1, t2] += 1
```

```python
print(bi)
```

```python
print(tri)
```

```python
model = {}
for (t0, t1, t2), freq in tri.items():
    if (t0, t1) in model:
        model[t0, t1].append((t2, freq/bi[t0, t1]))
    else:
        model[t0, t1] = [(t2, freq/bi[t0, t1])]
    
```

```python
print(model)
```

```python
lenght = 10
prefix = 'кошка'
phrase = ''
if not prefix:
    t0, t1 = choice(list(filter(lambda key: key[0] == prefix, model)))
for i in range(lenght):
    phrase += ' ' + t1
    t0, t1 = t1, choice(model[t0, t1])[0]

print(phrase.capitalize())
```
